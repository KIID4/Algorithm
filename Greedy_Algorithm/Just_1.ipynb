```python
n, k = map(int, input().split())
count = 0

while True:
    num = (n // k) * k # n에서 k로 나눌 수 있는 최대의 수
    count += (n - num) # 1을 만들어야 하므로 - 만큼 횟수증가(ex) n = 25, k = 3 => 25 - 24 = 1)
    n = num # 나누어 떨어지는 숫자

    if n < k: # 더 이상 나눌 수가 없을 때
        break

    count += 1 # 나누기 연산 수행
    n = n // k

count -= 1 # 항상 0으로 떨어지지만 문제에선 1이 될 때까지임
print(count)
```

### 핵심
  - n의 수를 k로 나눌 수 있는 최소횟수를 구할 땐 나눌 수 있는 만큼 먼저 나눈 후 후처리 진행
  - 조건에 맞춰서 0으로 나눠지는게 아니라 1로 나누어 져야하므로 횟수를 1번 차감해야함
